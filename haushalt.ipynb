```python
import pandas as pd
```

```python
df = pd.read_csv('Haushalt_2016_17.csv', sep=',', keep_default_na=False)
df
```

```python
import re

#Remove all product IDs
def remove_num(datum):
    return re.sub(r'(\d+\.?)+\s', '', datum)

remove_num('1.100.1.1.1.17 Rechnungsprüfung')
df['Produkt'] = df['Produkt'].apply(remove_num)
df['Teilprodukt'] = df['Teilprodukt'].apply(remove_num)

#Remove OE ID and redundant abbreviation

def remove_oe_num(datum):
    return re.sub(r'\(?\d+\)?\s', '', datum)

remove_oe_num('(10) FB Personal und Organisation')
df['OE'] = df['OE'].apply(remove_oe_num)
df['Kostenartengruppe'] = df['Kostenartengruppe'].apply(remove_oe_num)
df['OE'] = df['OE'].map(lambda x: x.lstrip('FB'))

#Add date, to be in accordance with the openspending form requirements
df['Datum'] = 2018
df
```

```python
#Delete duplicates

df.drop_duplicates(inplace=True)
```

```python
#Produce NaNs so that Pandas can replace these values
import numpy as np
df.replace('',np.nan,inplace=True)

#Getting the Organisationseinheiten, so that we have something to feed the fill forward function
organisations_einheiten = df.OE
organisations_einheiten = organisations_einheiten.dropna()

df['OE'] = df['OE'].apply(df.fillna(value=organisations_einheiten, method='ffill',axis=1))
```

```python
#There are summatations for one Produkt. They don't have a value in Kostenartengruppe. That is why we delete them.
df[~pd.isnull(df.Kostenartengruppe)]
```

```python
df.to_csv('haushalt_clean.csv', index=False)
```
