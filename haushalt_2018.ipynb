```python
import pandas as pd
import re
import goodtables
```

```python
df = pd.read_csv('Haushaltsentwurf_2018.csv', sep=';', keep_default_na=False)
df.head()
```

```python
# Remove all product IDs.
num_regex = r'(\d+\.?)+\s'
df['Produkt'] = df['Produkt'].replace(regex=num_regex, value='')
df['Teilprodukt'] = df['Teilprodukt'].replace(regex=num_regex, value='')

# Remove OE ID and redundant abbreviation.
oe_num_regex = r'\(?\d+\)?\s'
df['OE'] = df['OE'].replace(regex=oe_num_regex, value='')
df['Kostenartengruppe'] = df['Kostenartengruppe'].replace(regex=oe_num_regex, value='')

# # Remove FB prefix.
df['OE'] = df['OE'].replace(regex=r'^FB\s', value='')


# Add date, to be in accordance with the openspending form requirements.
df['Datum'] = 2018
df.head()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df = df.rename(columns={'OE': 'Organisationseinheit', 
                        'Interne Leist.': 'Interne Leistung',
                        '2018 (Plan)': 'Betrag'})
df.head()
```

```python
save_filename = 'haushalt_clean.csv' 
df.to_csv(save_filename, index=False)
```

```python
report = goodtables.validate(save_filename, row_limit=20_000)
report
```
